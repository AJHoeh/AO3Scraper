```python
url = 'http://archiveofourown.org/works/5937274?view_adult=true&amp;view_full_work=true'
headers = {'user-agent' : 
       'Chrome/52 (Macintosh; Intel Mac OS X 10_10_5); Jingyi Li/UC Berkeley/noon@berkeley.edu'}
req = requests.get(url, headers=headers)
src = req.text
#print src
soup = BeautifulSoup(src, 'html.parser')
```

```python
def get_tag_info(category, meta):
    '''
    given a category and a 'work meta group, returns a list of tags (eg, 'rating' -> 'explicit')
    '''
    tag_list = meta.find("dd", class_=str(category) + ' tags').find_all(class_="tag")
    return [result.text for result in tag_list] 
    
```

```python
def get_stats(meta):
    '''
    returns a list of  
    language, published, completed, words, chapters, comments, kudos, bookmarks, hits
    '''
    language = meta.find("dd", class_="language").text.rstrip().lstrip()
    published = meta.find("dd", class_="published").text
    completed = meta.find("dd", class_="status").text
    words = meta.find("dd", class_="words").text
    chapters = meta.find("dd", class_="chapters").text
    comments = meta.find("dd", class_="comments").text
    kudos = meta.find("dd", class_="kudos").text
    bookmarks = meta.find("dd", class_="bookmarks").text
    hits = meta.find("dd", class_="hits").text
    return [language, published, completed, words, chapters, comments, kudos, bookmarks, hits]
```

```python
meta = soup.find("dl", class_="work meta group")

rating = get_tag_info("rating", meta)
category = get_tag_info("category", meta)
fandom = get_tag_info("fandom", meta)
pairing = get_tag_info("relationship", meta)
characters = get_tag_info("character", meta)
additional_tags = get_tag_info("freeform", meta)

print ("rating: ", rating)
print ("c: ", category)
print ("f: ", fandom)
print ("p: ", pairing)
print ("char: ", characters)
print ("add: ", additional_tags)

print ("stats: ", get_stats(meta))
```
